```python
import pandas as pd
import numpy as np
# tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('./data/sonar3.csv', header=None)
```

```python
df.head()
```

```python
# 일반 암석과 광석의 숫자 (1: 광석, 0: 일반 암석)
# 111개 광석, 97개 일반 암석, 총 208개
df[60].value_counts()
```

```python
# 변수와 결과 분리
X = df.iloc[:, 0:60]
y = df.iloc[:, 60]
```

```python
# 딥러닝 모델 구성 및 학습
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # binary classification

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history = model.fit(X, y, epochs=200, batch_size=10)
```

```python
# (대부분) 정확도 1.000 (100%) 모델 학습 완료
result = (model.predict(X) > 0.5).reshape(-1)
sum(result == y) # 208개 일치
```

```python
# 과적합(overfitting) 대응을 위해 학습셋과 테스트 셋으로 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=True)
    
# 208개 중 30%를 테스트셋으로 사용 => 학습셋 145개 / 테스트 셋 63개
np.shape(X_train), np.shape(X_test), np.shape(y_train), np.shape(y_test)
```

```python
# 테스트셋의 결과
score = model.evaluate(X_test, y_test)
score # loss, accuracy
```

```python
# 학습셋과 테스트셋으로 분리한 내용으로 모델 구성 및 학습
# 딥러닝 모델 구성 및 학습
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # binary classification

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history = model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 학습의 결과
training_set_score = model.evaluate(X_train, y_train)
# 테스트셋의 결과
test_set_score = model.evaluate(X_test, y_test)

# 테스트 셋의 정확도가 학습셋의 정확도보다 많이 낮은 것을 확인할 수 있음
print('training set score(loss, accuracy)', training_set_score)
print('test set score(loss, accuracy)', test_set_score)
```

```python
# 모델 저장과 재사용(파일 포맷은 다양하게 있음, 책에서는 hdf5 사용)
model.save('./model/ch13/my_model.hdf5')
del model # 메모리에서 삭제
```

```python
# K겹 교차 검증k-fold cross validation
from sklearn.model_selection import KFold

# 5개 fold 구분
k = 5 
# shuffle로 데이터를 섞어셔 folding
kfold = KFold(n_splits=k, shuffle=True) 

# 결과 저장을 위한 array
acc_score = []

# kfold에 따라 데이터 셋 분리한 각 iteration 모델 만들어서 시행
for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        loss='binary_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)
    
    acc = model.evaluate(X_test, y_test)[1]
    acc_score.append(acc)
```

```python
print('정확도:', acc_score)
print('정확도 평균:', np.mean(acc_score))
```
